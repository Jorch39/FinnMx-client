### Preparación de datos para el modelo

```python
#Separamos las variables 
data_train = fin[['income', 'age', 'sex']]
data_train.head()
```

```python
#Definimos nuestra variable objetivo 
data_target= fin['invest']
data_target.head()
```

```python
print(data_train.iloc[50,:])
print('invest:', data_target.iloc[50])
```

```python
#Corroboramos que sean valores numéricos 
fin.dtypes
```

### Creamos el modelo

```python
# Separamos el dataset en un conjunto de entrenamiento y de prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_train, data_target, test_size=0.2, random_state=0)
```

```python
#Definimos arbol de decisión como modelo 
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
#Visualización del modelo 
from sklearn.tree import export_graphviz
from IPython.display import Image 
import io 

dot_data = io.StringIO()
export_graphviz(clf, out_file=dot_data, filled=True, rounded=True, special_characters=True, feature_names= ['income', 'age', 'sex'])
graph = pydotplus.graphviz.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

### Ajustamos el modelo 

```python
#Checamos que tan puros son los datos
clf = tree.DecisionTreeClassifier(criterion='gini', max_depth=3, min_samples_leaf=5)
clf = clf.fit(X_train, y_train)
```

```python
dot_data = io.StringIO()
export_graphviz(clf, out_file=dot_data, filled=True, rounded=True, special_characters=True, feature_names= ['income', 'age', 'sex'])
graph = pydotplus.graphviz.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

#### Con esto sacamos las reglas de negocio que necesitamos para llevarlos a nuestra app 

### Testeamos resultados

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
y_pred = clf.predict(X_train)
train_accuracy = accuracy_score(y_train, y_pred)
f'Precisión en el conjunto de entrenamiento: {train_accuracy}'
```

```python
y_pred = clf.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
f'Precisión en el conjunto de prueba: {test_accuracy}'
```
